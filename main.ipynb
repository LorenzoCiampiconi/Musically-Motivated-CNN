```python
import data_handler as dh 
import pandas as pd
import tensorflow as tf
```

set variables 

```python
DATA_PATH = "dataset/"
DATA_NAME = "train-mel.arff"
```

import data

```python
data = dh.import_arff(DATA_PATH + DATA_NAME)


data['class'] = pd.Categorical(data['class'])
data['class'] = data['class'].cat.codes

data.head()
```

convert dataset from pandas to tensorflow dataset

```python
target = data.pop("class")
target_hoe = pd.get_dummies(target)
dataset = tf.data.Dataset.from_tensor_slices((data.values, target_hoe.values))
```

```python
for feat, targ in dataset.take(5):
  print ('Features: {}, Target: {}'.format(feat, targ))
```

```python
train_dataset = dataset.shuffle(len(data)).batch(1)
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
  ])

  model.compile(optimizer='adam',
                loss='categorical_crossentropy',
                metrics=['accuracy'])
  return model
```

```python
model = get_compiled_model()
model.fit(train_dataset, epochs=10)
```
